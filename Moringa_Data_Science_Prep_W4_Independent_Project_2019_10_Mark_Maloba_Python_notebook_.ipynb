# Cleaning

```
#The future stations do not help with our analysis of current events, so the related "Scheduled at" column, by extension, isn't necessary. 
#That and the comments column also happen to be the 2most incomplete columns as seen below:
AUTOdf.info()
```

```
#Hence we can drop them
#AUTOdf.drop(["Displayed comment","Scheduled at"], axis = 1, inplace= True)
AUTOdf.info()
```

```
#Computing IQR for each column to detect Outliers

Q1 = AUTOdf.quantile(0.25)
Q3 = AUTOdf.quantile(0.75)
IQR = Q3 - Q1

print(IQR)
```

```
#Removing the outliers

print((AUTOdf  < (Q1 - 1.5 * IQR)) |(AUTOdf > (Q3 + 1.5 * IQR)))

AUTOdf2 = AUTOdf[~((AUTOdf < (Q1 - 1.5 * IQR)) |(AUTOdf > (Q3 + 1.5 * IQR))).any(axis=1)]
AUTOdf2.shape
```

```
#Combining Date & Time in preparation for Analysis
#We will also drop the unnecessary split date & time columns

# AUTOdf2['Date'] = AUTOdf2["year"].map(str) +':'+ AUTOdf2["month"].map(str) +':'+ AUTOdf2["day"].map(str)
# AUTOdf2.head()
# AUTOdf2['Time']=AUTOdf2["hour"].map(str) +':'+ AUTOdf2["minute"].map(str)
# AUTOdf2.head()
# AUTOdf2['Date.Time']= AUTOdf2["Date"].map(str) +':'+ AUTOdf2["Time"].map(str)
# AUTOdf2.head()
AUTOdf2.drop(['year','month','day','minute','Date','Time'],axis=1,inplace=True)
AUTOdf2.head()
```

```
#Last cleaning action is to check for & deal with duplicate values (if any) in our DataFrame

AUTOdf2.duplicated().sum()
```

# Analysis

```
#Subsetting: Separate Bluecar activity for Paris in the month of April
AUTOdf3 = AUTOdf2[(AUTOdf2['City']=='Paris') & (AUTOdf2['Bluecar counter']!=0)]
AUTOdf3.head(5)
```

```
#Grouping the data by station

AUTOdf4 = AUTOdf3.sort_values(['Date.Time']).groupby(['ID','Date.Time']).sum()
AUTOdf4.head(15)
```

```
#Custom function to indicate car pickup vs return. +ve = pickup, -ve = return 

AUTOdf4['rt_or_pick'] = AUTOdf4.groupby(['ID'])['Bluecar counter'].apply(lambda x:  x.shift(1) - x)
AUTOdf4['rt_or_pick']
AUTOdf4.head(15)
```
